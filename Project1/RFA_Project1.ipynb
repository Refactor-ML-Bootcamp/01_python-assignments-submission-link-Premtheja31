finding outliers and removing

```python
def removing_outliers(data,X):
  q25,q75=np.percentile(data[X],[25,75])
  iqr=q75-q25
  min_o=q25-(1.5*iqr)
  max_o=q75+(1.5*iqr)
  ad=[x for x in data[X] if x<=max_o and x>=min_o]
  return ad
  
```
